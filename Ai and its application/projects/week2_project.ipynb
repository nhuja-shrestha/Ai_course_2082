#2. DATA QUALITY CHECK

```python
#checking if there are any missing values or not
print("\nMissing Values:")
missing_values=df_copy.isnull().sum()
print(missing_values)
```

```python
#If there are any missing values we would handle it like:
#first seperating numerical comulns and categorical column
missing_values=df_copy.isnull().sum()
num_cols = df_copy.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df_copy.select_dtypes(include=['object']).columns
if missing_values.sum() > 0 :
  #numerical columns lai median le fill garne
    for col in num_cols:
        if df_copy[col].isnull().sum() > 0:
            median_val = df_copy[col].median()
            df_copy[col].fillna(median_val, inplace=True)
            print(f"Filled {col} with median: {median_val}")

    # Fill categorical with mode
    for col in cat_cols:
      if df_copy[col].isnull().sum() > 0:
            mode_val = df_copy[col].mode()[0]
            df_copy[col].fillna(mode_val, inplace=True)
            print(f"Filled {col} with mode: {mode_val}")
else:
  print("\nNo missing values")
```

#3. Outlier detection and handling

```python
# Using boxplot to detect outliers in Monthly Income
# defining figsize for a user demanded size for clearer view

sns.boxplot(x=df_copy['MonthlyIncome'])
plt.title("Boxplot of Monthly Income")
plt.show()
```

```python
#using IQR method to handle outliers
#IQR= Inter Quartile Range
q1 = df_copy['MonthlyIncome'].quantile(0.25)
q3 = df_copy['MonthlyIncome'].quantile(0.75)
iqr = q3 - q1
df_filtered = df_copy[(df_copy['MonthlyIncome'] >= q1 - 1.5 * iqr) & (df_copy['MonthlyIncome'] <= q3 + 1.5 * iqr)]

sns.boxplot(x=df_filtered['MonthlyIncome'])
plt.title("Boxplot of monthly Income with less outliers")
plt.show()
```

```python
#using IQR method again to remove existing outliers
#IQR= Inter Quartile Range
q1 = df_filtered['MonthlyIncome'].quantile(0.25)
q3 = df_filtered['MonthlyIncome'].quantile(0.75)
iqr = q3 - q1
df_filtered1 = df_filtered[(df_filtered['MonthlyIncome'] >= q1 - 1.5 * iqr) & (df_filtered['MonthlyIncome'] <= q3 + 1.5 * iqr)]

sns.boxplot(x=df_filtered1['MonthlyIncome'])
plt.title("Boxplot of monthly Income with lesser outliers")
plt.show()
#After using IQR we removed most of the outliers. There are still some of them left because they are within acceptable range
#please mention this on our presentation slides.
```

##4.3 Multivarriate

```python
#using heatmap for multivariate data analysis
#label encoding on attrition vanlue
#1 vaneko yes i.e company xodera gakoo
#0 vaneko no i.e company ma ajjai xann
df_copy['Attrition_n'] = df_copy['Attrition'].map({'Yes': 1, 'No': 0})
# Selecting most relevant feature that can affect attrition only
features = [
    'Age', 'MonthlyIncome','TotalWorkingYears',
    'YearsAtCompany', 'YearsInCurrentRole','Attrition_n']
corr=df_copy[features].corr()
# defining figsize for a user demanded size for clearer view
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')      #fmt to display float values only two decimal value
plt.title("Correlation Heatmap")
plt.show()
```

#Creating bins for Age

```python
#seeing how attrition is affected with age group
df_copy['AgeGroup'] = pd.cut(df_copy['Age'], bins=[18, 30, 40, 50, 60],
                             labels=['18-30', '31-40', '41-50', '51-60'])
sns.countplot(x='AgeGroup',hue='Attrition',data=df_copy)
plt.title('Attrition by Age Group')
plt.show()
```
